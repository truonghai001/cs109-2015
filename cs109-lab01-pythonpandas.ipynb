#### Cleaning

```python
df.dtypes
```

```python
# check to see if there is null values in years serie
df[df.year.isnull()]
```

```python
df = df[df.year.notnull()]
df.shape
```

```python
df['rating_count'] = df.rating_count.astype(int)
df['review_count'] = df.review_count.astype(int)
df['year'] = df.year.astype(int)
df.dtypes
```

#### Visualizing

```python
df.ratings.hist()
```

```python
sns.set_context("notebook")
meanrat = df.ratings.mean()
print(meanrat, np.mean(df.ratings), df.ratings.mean())
with sns.axes_style('whitegrid'):
    df.ratings.hist(bins=30, alpha=0.4)
    plt.axvline(meanrat, 0, 0.75, color='r', label='Mean')
    plt.xlabel('average rating of book')
    plt.ylabel('Coutns')
    plt.title('Ratings Histogram')
    plt.legend()
```

```python
df.review_count.hist(bins=np.arange(0, 40000, 400))
```

We make a scatterplot in matplotlib of rating against year.
By setting the alpha transparency low, we can see how the density of highly rated books on goodreads has changed.

```python
plt.scatter(df.year, df.ratings, lw=0, alpha=0.08)
plt.xlim([1900, 2010])
plt.xlabel('Year')
plt.ylabel('Rating')
```
